```python
# importing the library
import numpy as np
import matplotlib.pyplot as plt

#increasing plot size for better visibility
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size
```

```python
# data to be plotted
x = ["'17","'18","'19","'20","'21"]
DRS = np.array([-5,-11,-5,-4,-13])
UZR = np.array([-4.6,-4.5,2.7,-5.6,-5.5])
OAA = np.array([-7,-7,17,-4,-13])
zeros = [0,0,0,0,0]
 
# plotting
plt.title("Rafael Devers Defensive Metrics")
plt.xlabel("Year")
plt.ylabel("Value")
plt.plot(x, DRS, color ="red", label = "DRS")
plt.plot(x, UZR, color ="green", label = "UZR")
plt.plot(x, OAA, color ="blue", label = "OAA")
plt.plot(x,zeros, color ="gray", linestyle='dashed')
plt.legend(loc="upper left")
plt.show()
```
